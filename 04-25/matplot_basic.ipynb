```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 2, 1]

plt.plot(x, y)
plt.title('sign graph')
plt.show()
```

```python


plt.plot(x,y)
plt.title('sign graph') #제목
plt.grid()

plt.xlabel('time')
plt.ylabel('amplitude') # 제목

plt.show()
```

```python
plt.plot(x,np.sin(x), label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.title('sin & cos graph')
plt.grid()

plt.show()
```

```python
plt.plot(x,np.sin(x), label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.title('sin & cos graph')

plt.grid()
plt.legend()  # label= 'sin'을쓰려면 legned를 써야 한다. 


plt.show()
```

```python
plt.plot(x,np.sin(x), label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.title('sin & cos graph')

plt.grid()
plt.legend(loc = (1.05 ,0.05))  # label= 'sin'을쓰려면 legned를 써야 한다. 


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.plot(x,np.sin(x), label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.title('sin & cos graph')

plt.grid()
plt.legend(loc = (1.05 ,0.05))  # label= 'sin'을쓰려면 legned를 써야 한다. 

plt.xlim(0, np.pi)
plt.ylim(-1.2,1.2)

#x.lim 
#y.lim


plt.show()
```

# 스타일(Style)

     https://matplotlib.org/stable/api/markers_api.html

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5,6]
y= [2,4,8,12,16,20]
```

```python
plt.plot(x,y, linewidth=5)  # 선굵기
```

```python
plt.plot(x,y, marker='o')  
```

```python
plt.plot(x,y, marker='*', markersize = 10, markeredgecolor='red', linestyle='-.')  
```
