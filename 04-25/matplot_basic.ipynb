```python
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import numpy as np

# make figure and assign axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)

# pie chart parameters
overall_ratios = [.27, .56, .17]
labels = ['Approve', 'Disapprove', 'Undecided']
explode = [0.1, 0, 0]
# rotate so that first wedge is split by the x-axis
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.1f%%', startangle=angle,
                     labels=labels, explode=explode)

# bar chart parameters
age_ratios = [.33, .54, .07, .06]
age_labels = ['Under 35', '35-49', '50-65', 'Over 65']
bottom = 1
width = .2

# Adding from the top matches the legend.
for j, (height, label) in enumerate(reversed([*zip(age_ratios, age_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label,
                 alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax2.set_title('Age of approvers')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

# use ConnectionPatch to draw lines between the two plots
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(age_ratios)

# draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(4)
ax2.add_artist(con)

# draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(4)

plt.show()
```

```python


## 3. 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리시오.
#연도별 평균 데이터를 구하는 코드는 다음과 같습니다.
#df_group = df.groupby('개봉 연도').mean()

# group the dataframe by '개봉 연도' and calculate the mean of '평점'
df_group = df.groupby('개봉 연도')['평점'].mean()

# create a line chart
plt.plot(df_group.index, df_group.values, marker='o')
plt.xlabel('개봉 연도')
plt.ylabel('평점')
plt.title('개봉 연도별 평점 변화 추이')
plt.show()
```

```python
## 3. 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리시오.
#연도별 평균 데이터를 구하는 코드는 다음과 같습니다.
df_group = df.groupby('개봉 연도').mean(numeric_only=True)

# '개봉 연도'를 그룹으로 하여 각 그룹별로 '평점' 열의 평균 값을 계산하고, 
# 그 결과를 새로운 데이터프레임(df_group)에 저장합니다. 
# numeric_only=True는 수치형 열만 계산하도록 하는 옵션입니다.


# df_group.index refers to the index of the df_group dataframe, which is the '개봉 연도' column. 
# In other words, it is a Pandas series object that contains the unique values of the '개봉 연도' column, 
# which are the years in which the movies were released.


plt.plot(df_group.index, df_group['평점'])


# df_group의 인덱스(개봉 연도)를 x축으로, '평점' 열의 평균 값을 y축으로 하는 꺾은선 그래프를 그립니다. 
# plt.plot()은 주어진 데이터를 꺾은선 그래프로 시각화하는 함수입니다.

# df_group은 groupby를 통해 개봉 연도로 그룹화된 데이터프레임입니다. 
# df_group.index는 해당 그룹화된 데이터프레임의 인덱스인 개봉 연도 정보를 반환합니다. 
# 따라서 plt.plot(df_group.index, df_group['평점']) 코드는 x축에 개봉 연도 정보를, 
# y축에는 해당 연도별 평균 평점 정보를 나타내는 꺾은선 그래프를 그리는 것입니다.


```

```python
## 5) 평점이 9점 이상인 영화의 비율을 확인할 수 있는 원 그래프를 제시된 세부 사항을 적용하여 그리시오.
#- label: 9점 이상 / 9점 미만
#- 퍼센트: 소수점 첫 번째 자리까지 표시
#- 범례: 그래프 우측에 표시

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 8
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    '영화': ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도': [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수': [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점': [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)

rating_counts = pd.cut(df['평점'], [0, 9, 10]).value_counts()
rating_percents = rating_counts / rating_counts.sum() * 100

plt.pie(rating_percents, labels=['9점 미만', '9점 이상'], autopct='%.1f%%')
plt.legend(loc='right')
plt.axis('equal')
plt.show()
```
