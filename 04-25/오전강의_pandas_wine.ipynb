- type: 와인의 종류 (white, red) 
- alcohol: 알콜 도수

```python
# 연습문제 2: 기술통계 정보를 출력해보세요.
df.describe()
```

```python
# 연습문제 3: 와인의 종류(quality)별로 데이터가 몇 개씩 있는지 출력해보세요.
# df['quality'].value_counts()
df['type'].value_counts()
```

```python
# 연습문제 7: 각 와인의 타입(type)별로 평균 알코올 도수(alcohol)와 
# 평균 당도(residual sugar)를 구해보세요.
# white, red의 특징을 좀 알고싶다.

df.groupby('type')['alcohol', 'residual sugar'].mean()
```

```python
# 연습문제 8: 'quality' 열의 값이 5 이상인 데이터만 추출하시오.

# quality가 5이상, type은 레드
good_quality_df = df[(df['quality'] >= 5) & (df['type'] == 'red')]
good_quality_df
```

```python
# 연습문제 9: 'fixed acidity' 열과 'pH' 열의 값이 각각 3 이상, 4이상 인 데이터만 추출하시오.
df[(df['fixed acidity'] >= 3) & (df['pH'] >= 4)]
```

```python
# 연습문제 10: 'volatile acidity' 열의 값이 0.5 이하인 데이터의 
# 'quality' 열의 평균을 계산하세요.

df[df['volatile acidity'] <= 0.5]['quality'].mean()
```

```python
# 연습문제 11: 'residual sugar' 열의 값이 2 이하인 데이터의 'density' 열의 중앙값을 계산하세요.

df[df['residual sugar'] <= 2]['density'].median()
```

```python
# 연습문제 12: 'alcohol' 열이 가장 높은 5개 데이터의 
# 'quality' 열 값을 출력하세요.

new_df = df.sort_values('alcohol', ascending=True).tail(10) # 오름차순 일 때의 하위 5개 (작은 숫자 => 큰 숫자로 숫자가 커짐)
new_df['alcohol']
```

```python
new_df = df.sort_values('alcohol', ascending=False).head(10) # 내림차순 했을 때 상위 5개 (큰 숫자 => 작은 숫자로 숫자가 작아짐)
new_df['alcohol']
```

```python
# 연습문제 13: 'chlorides'(염화물) 열의 값이 가장 높은 데이터 10개의 'pH' 열 값을 출력하세요.

df.sort_values('chlorides', ascending=False).head(10)['pH']
```

```python
# 연습문제 14: 'sulphates' 열의 값이 0.4 이하인 데이터의 'quality' 열 값을 1로 변경하세요.

# df[df['sulphates'] <= 0.4]['quality'] = 1
# df.loc[df["sulphates"] <= 0.4, "quality"] = 1
# df[df['sulphates'] <= 0.4]['quality']

index = df['sulphates'] <= 0.4 # 인덱싱 조건
```

```python
# 연습문제 15: 
# 'fixed acidity' 열의 값이 7 이하인 데이터의 'quality' 열 값을 0으로, 
# 'fixed acidity' 열의 값이 8 이상인 데이터의 'quality' 열 값을 2로 변경하세요.
# 나머지 데이터는 변경하지 않습니다.

low_acidity_df = df[df['fixed acidity'] <= 7]
low_acidity_df.loc[:, 'quality'] = 0 # quality 컬럼 값을 0으로 변경
low_acidity_df

high_acidity_df = df[df['fixed acidity'] >= 8]
high_acidity_df.loc[:, 'quality'] = 2
high_acidity_df

# categorizing 작업을 해주기 위해 사용합니다. => labmda
# 우리 고객이 20~50대까지 존재해. 
# 고객을 타겟팅해서 광고하고 싶어. 여러분들은 어떻게 해야하죠?
# 고객 타겟팅 => 고객을 그룹화 (나이 기준)

# 나이를 5살 기준으로 할 수도 있구요.
# 20~25
# 26~30
# 31~35
```

```python
concat_data_df = pd.concat([low_acidity_df, high_acidity_df], ignore_index=True) 
concat_data_df['quality'].value_counts()
```

```python
# 연습문제 16: 'quality' 열의 값이 0인 데이터 중 'fixed acidity' 열의 값을 7로 변경하세요.

zero_quality_df = concat_data_df[concat_data_df['quality'] == 0]
zero_quality_df.loc[:, 'fixed acidity'] = 7
zero_quality_df
```

```python
# 연습문제 17: 'quality' 열의 값이 6 이하인 데이터 중 'pH' 열의 값을 3으로 변경하세요.

low_quality_df2 = concat_data_df[concat_data_df['quality'] <= 6]
low_quality_df2.loc[:, 'pH'] = 3
low_quality_df2
```
