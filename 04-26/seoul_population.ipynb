(4) 각 구별 전체 인구를 이용해서 구별 '외국인비율'과 '고령자비율' 컬럼을 추가하고, 소수점 첫째자리 반올림하기

- 외국인 비율: 외국인 / 인구수 * 100

- 고령자 비율: 고령자 / 인구수 * 100

```python
new_df.info()

#부비트랩이었다. 
```

```python
new_df.info()

new_df['인구수'] = new_df['인구수'].astype('int')
new_df['한국인'] = new_df['한국인'].astype('int')
new_df['외국인'] = new_df['외국인'].astype('int')
new_df['고령자'] = new_df['고령자'].astype('int')


new_df.info()
```

```python
import numpy as np

new_df['외국인비율'] = np.round(new_df['외국인'] / new_df['인구수'] * 100, 1)
new_df['고령자비율'] = np.round(new_df['고령자'] / new_df['인구수'] * 100, 1)

new_df
```

# (9) 구별 인구수 데이터를 바 그래프로 내림차순 정렬 후 시각화하기

```python
### (9) 구별 인구수 데이터를 바 그래프로 내림차순 정렬 후 시각화하기
plt.rcParams['font.family']="Malgun Gothic" # windows
# plt.rcParams['font.family']="AppleGothic" # mac
# visual_df['인구수'].plot(kind='bar', grid=True)
visual_df['인구수'].sort_values().plot(kind='bar', grid=True) # 오름차순 정렬을 해준다

# df.sort_values(by='인구수')
# df['인구수'].sort_values()
```

- 판다스로 데이터 전처리

- Sikit-learn, Pytorch, Tensorflow => 머신러닝, 딥러닝 => 예측 (+알고리즘, ...)

```python

x = visual_df.index
y = visual_df['인구수']

plt.figure(figsize=(16,10))
plt.title("서울시 구별 인구 현황")

plt.xlabel("gu")
plt.ylabel("population")
plt.bar(x,y)

plt.grid()
plt.bar(x,y)

# plt.show() show를 하면 savefig가 잘 안먹는 현상이 있다. 
plt.savefig("서울시인구현황.png")
```
