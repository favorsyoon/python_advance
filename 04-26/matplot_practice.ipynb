```python
# matplotlib를 사용하여 0부터 10까지의 정수를 x축으로, 해당 정수의 제곱을 y축으로 하는 선 그래프를 그려보세요.

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11)
y = x ** 2
print(x, y)

plt.plot(x, y)
plt.show()
```

```python
# matplotlib를 사용하여 -10부터 10까지의 범위에서 sin(x) 함수를 그려보세요.

x = np.arange(-10, 10) # time
y = np.sin(x) # amplitude

plt.plot(x,y)
```

```python
# matplotlib를 사용하여 0부터 2파이까지의 범위에서 sin(x)과 cos(x) 함수를 함께 그려보세요.
x = np.linspace(0, 2 * np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos")
plt.legend()
plt.show()
```

```python
# matplotlib를 사용하여 0부터 1까지의 범위에서 x^2, x^3, x^4 함수를 함께 그려보세요.

x = np.linspace(0, 1, 100)

y1 = x ** 2
y2 = x ** 3
y3 = x ** 4

plt.plot(x, y1, label="$x^2$")
plt.plot(x, y2, label="$x^3$")
plt.plot(x, y3, label="$x^4$")

plt.legend()
plt.show()
```

```python
#matplotlib를 사용하여 산포도(scatter plot)를 그려보세요. 
#x축은 1부터 10까지의 정수, y축은 해당 정수의 제곱으로 설정하세요.

x = np.arange(1, 11,) # 1~10
y = x ** 2

plt.scatter(x, y)
plt.show()
```

```python
# matplotlib를 사용하여 막대 그래프(bar chart)를 그려보세요.
# x축은 ['A', 'B', 'C', 'D', 'E'], y축은 각 항목의 개수로 설정하세요.

x = ['A', 'B', 'C', 'D', 'E']
y = [10, 7, 8, 9, 5] # height

plt.bar(x,y)
plt.show()
```

```python
#matplotlib를 사용하여 원 그래프(pie chart)를 그려보세요. 
#각 항목의 비율을 적절하게 조절하여 보기 좋게 만드세요.

languages = ['python', 'C++', 'C#', 'Java']
market = [60, 10, 5, 30]
explode = [0.05] * 4

plt.pie(market, labels=languages, autopct="%1.f%%", explode=explode)
plt.show()
```

```python
#matplotlib를 사용하여 히스토그램(histogram)을 그려보세요. 
#0부터 100까지의 범위에서 무작위로 1000개의 정수를 생성하여 이를 그립니다.

data = np.random.randint(1,100, 1000)
print(len(data))

plt.hist(data, bins=20)
```

```python
#matplotlib를 사용하여 다중 서브플롯(multi subplot)을 그려보세요. 
#1행 2열의 구조에서, 왼쪽 서브플롯에는 0부터 10까지의 범위에서 sin(x) 함수를, 
#오른쪽 서브플롯에는 0부터 10까지의 범위에서 cos(x) 함수를 그립니다.

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,4))

ax[0].plot(x, y1)
ax[1].plot(x, y2)

plt.show()
```
