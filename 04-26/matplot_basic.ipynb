```python
import numpy as np

x = np.arange(0, 12, 0.01)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.plot(x,y-1)
plt.plot(x,y-2)

plt.title("sign graph") # 제목
plt.grid() # 눈금, 격자

plt.xlabel("time") # 시간
plt.ylabel("amplitude") # 진폭

plt.show()
```

```python
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")

plt.grid()
plt.legend(loc=(1.05,0.05))

plt.title("sin & cos graph")
plt.show()
```

## 스타일 (Style)

```python
x = [1,2,3,4,5,6]
y = [2,4,8,12,16,20]
```

```python
plt.figure(facecolor='red')

# plt.plot(x,y, linewidth=10)
plt.plot(x,y, linewidth=3, marker='d', markersize=30,
         markeredgecolor='red', linestyle="-.")

plt.figure(figsize=(60,30), dpi=800)
plt.show()
 # graph size 조절할 수 있도록 도와주는 함수. # dpi: dots per inch
```

```python
x_value = np.arange(1, 20, 2)
y_value = 2 * x_value

plt.plot(x_value, y_value, color='red', marker='o', markersize=6, linestyle='dashed', linewidth=2, label='line') # 선그래프 
plt.bar(x_value, y_value, color='pink', label='bar') # 막대 그래프

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.title("growth report")

plt.show()
```

```python
x_value01 = np.arange(1, 10)
x_value02 = np.arange(1, 20)

y_value01 = 2 * x_value01
y_value02 = 2 * x_value02

fig, (ax_01, ax_02) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

ax_01.plot(x_value01, y_value01, color='red', marker='o', markersize=6, linestyle='dashed', linewidth=2, label='line')
ax_02.bar(x_value02, y_value02, color='green', label='bar')

ax_01.set_xlabel("ax_01 x label")
ax_02.set_xlabel("ax_02 x label")

ax_01.legend()
ax_02.legend()

plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 6)) # subplot은 하나씩 설정, subplots은 한번에 설정.

ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6, label='temp_01')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='temp_02')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=6, label='temp_03')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='temp_04')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend() 
ax[1][0].legend()
ax[1][1].legend() 

#plt.legend()
plt.show()
```

### 산점도 그래프 (Scatter)

- 여러 집단 간의 비교를 위해 그린다.

```python
x1 = np.random.rand(100) - 1.0
x2 = np.random.rand(100)
y1 = np.random.rand(100)
y2 = np.random.rand(100)

plt.scatter(x1, y1, marker="+")
plt.scatter(x2, y2, marker="x")
```

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('http://raw.githubusercontent.com/datagy/data/main/KNN_data.csv')
df.head()
df.plot()
```

```python
# Label => Large, Medium, Small 각각 df를 만드시오.

small_df = df[df['Label'] == "Small"]
medium_df = df[df['Label'] == "Medium"]
large_df = df[df['Label'] == "Large"]

ax = small_df.plot(x='x', y='y', kind='scatter', color='r', label="Small")
medium_df.plot(x='x', y='y', kind='scatter', color='g', label="Medium", ax=ax)
large_df.plot(x='x', y='y', kind='scatter', color='b', label="Large", ax=ax)

# plt.scatter(medium_df["x"], medium_df["y"], color= "violet", marker="v")
```

```python
small_df = df[df['Label'] == "Small"]
medium_df = df[df['Label'] == "Medium"]
large_df = df[df['Label'] == "Large"]

plt.scatter(small_df["x"], small_df["y"], color="r", label="Small")
plt.scatter(medium_df["x"], medium_df["y"], color="g", label="Medium")
plt.scatter(large_df["x"], large_df["y"], color="b", label="Large")

plt.legend()
plt.show()
```
